```python
datos = df1['Year']
intervalos = range(min(datos), max(datos) + 2)

plt.hist(datos, bins=8, color='#F2AB6D', rwidth=0.85)
plt.xticks(intervalos)
plt.title('Cantidad de terremotos por año en el mundo')
plt.ylabel('ID_Earthquake')
plt.xlabel('Year')

plt.show()
```

```python
data = df1['Year']
bins = (170, 10, 250)

plt.hist(data, bins = bins, density = True)
plt.title('Cantidad de terremotos por año')
plt.ylabel('Terremotos')
plt.xlabel('Year')

plt.show()
```
